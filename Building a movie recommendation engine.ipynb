```python
import pandas as pd
import numpy as np
import time
from datetime import datetime
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies.genres = movies.genres.str.split('|')
```

```python
# Create a genres list
i_list = []
for i in movies['genres']:
    for j in i:
        i_list.append(j)
genres_list = set(i_list)
genres_list
```

```python
for j in genres_list:
    movies[j] = 0
for i in range(movies.shape[0]):
    for j in genres_list:
        if (j in movies['genres'].iloc[i]):
            movies.loc[i,j] = 1
```

```python
movies.head(5)
```

```python
movies_1 = movies.iloc[:,3:]
movies_1
```

```python
# Normalize each row
movies_1['norm'] = np.sqrt(np.sum(movies_1.iloc[:,0:],axis=1))
movies_1.head(5)
```

```python
for i in movies_1.columns:
    movies_1[i] = movies_1[i] / movies_1['norm']
```

```python
movies_1.head(5)
```

```python
movies_1.drop(['norm'],axis=1,inplace=True)
movies_1.head(5)
```

```python
## M dot MT for the similarity
similarity_matrix = np.dot(movies_1,movies_1.T)
```

```python
similarity_matrix
```

```python
df_sm = pd.DataFrame(similarity_matrix, index=movies['movieId'], columns=movies['movieId'])
df_sm
```

```python
def gettopksimilar(moid,min_s,max_s,top_k=5,min_rat=4.1,min_t_rat=5,fyear='2010',tyear='2017'):
    
    ss = df_sm[df_sm.index==moid].T.sort_values(moid,ascending=False)
    ss1 = ss.reset_index()
    ss2 = ss1.rename(columns={'index':'movieId'})
    movies_0 = movies.iloc[:,:3]
    comb1 = pd.merge(ss2,movies_0,how='left',on='movieId')
    
    s = "{}-01-01".format(fyear)
    e = "{}-12-31".format(tyear)
    date_s = datetime.strptime(s,"%Y-%m-%d")
    timestamp_s = datetime.timestamp(date_s)
    date_e = datetime.strptime(e,"%Y-%m-%d")
    timestamp_e = datetime.timestamp(date_e)
    
    sel_date = ratings.loc[(ratings['timestamp']>=timestamp_s)&(ratings['timestamp']<=timestamp_e)]
    get_mov_id = sel_date.groupby(['movieId']).agg({'userId':'count','rating':'mean'}).reset_index()
    comb2 = pd.merge(comb1, get_mov_id,how='left',on='movieId')
    get_mov_id_1 = comb2.loc[((comb2[moid]>=min_s)&(comb2[moid]<=max_s)&(comb2['movieId']!=moid))&\
                           (comb2['userId']>min_t_rat)&(comb2['rating']>min_rat)].head(top_k)
    
    get_mov_id_1.columns = ['movieId','Similarity_Score','Recommended_Movie','Genres','total_rating','avg_rating']
    return get_mov_id_1
```

```python
gettopksimilar(5,0.3,1,10,3.5,20,'2010','2017')
```
